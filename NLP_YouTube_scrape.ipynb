This code searches youtube video by keyword using search().list method from YouTube data API v3. After searching the video, the results have their according unique video IDs, published date, time, and the title of videos. Then, it scrapes the videos comments using CommentThreads.list() method.                                                                             

```python
import googleapiclient.discovery
import pandas as pd
import googleapiclient.errors

api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = "Replace it with your KEY"

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=DEVELOPER_KEY)
```

Search 10 videos by keword "Trump"
You can modify the keyword by `q` variable. The number of maximum results in this code is 10. You can adjust this by `maxResults` variable.

```python
request = youtube.search().list(
    part="id,snippet",
    type="video",
    q="Impact of AI",
    videoDuration='short',
    videoDefinition='high',
    maxResults=40
)

response = request.execute()
print(response)
```

## Video information processing
Leave only necessary data and convert it to dataframe format

```python
for item in response['items']:
  # print(item['id']['videoId'])
  print(item['id']['videoId'], item['snippet']['publishedAt'], item['snippet']['title'], item['snippet']['publishTime'])

snippets = []

for item in response['items']:
  snippet = [item['id']['videoId'], item['snippet']['publishedAt'], item['snippet']['title'],item['snippet']['publishTime']]
  snippets.append([
      snippet[0],
      snippet[1],
      snippet[2],
      snippet[3]
  ])

df = pd.DataFrame(snippets, columns=['videoID', 'published_at', 'title', 'publishTime'])
df
```

## Scrape the comments by videoID
use `max_results` parameter to change the maximum number of comments(up to 100)

```python
list_comments = []
max_results = 100
for i in df['videoID']:
  # print(i)
  try:
    request_vid = youtube.commentThreads().list(part='snippet', videoId=i, maxResults=max_results, order='relevance')
    response_vid = request_vid.execute()

    comments = []

    for item in response_vid['items']:
      comment = item['snippet']['topLevelComment']['snippet']
      list_comments.append([
        i,
        comment['authorDisplayName'],
        comment['publishedAt'],
        comment['updatedAt'],
        comment['likeCount'],
        comment['textDisplay']
      ])

    # list_comments.append(comments)

  except:
      pass
print("The average number of videos with comments: ", len(list_comments)/max_results)
print("The number of total comments:", len(list_comments))
```

```python
df_comments = pd.DataFrame(list_comments, columns=['videoID', 'author', 'published_at', 'updated_at', 'like_count', 'text'])
df_comments
```
